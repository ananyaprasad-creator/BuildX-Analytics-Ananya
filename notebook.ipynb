```python
import pandas as pd
import os
print(os.listdir(path))
df = pd.read_csv(f"{path}/city_day.csv")
print(df.head())
```

```python
import pandas as pd


print("Initial Data Preview:")
print(df.head())
print("\nMissing values per column:")
print(df.isnull().sum())


df['PM2.5'] = df['PM2.5'].fillna(0)
df['PM10'] = df['PM10'].fillna(0)
df['AQI'] = df['AQI'].fillna(0)
df = df.dropna(subset=['City'])

df = df.drop_duplicates()
df['City'] = df['City'].str.strip().str.title()
df['AQI_Bucket'] = df['AQI_Bucket'].str.strip().str.title()


df['Date'] = pd.to_datetime(df['Date'])
df['AQI'] = pd.to_numeric(df['AQI'])


df = df.rename(columns={
    'city': 'City',
    'date': 'Date',
    'AQI_bucket': 'Pollution_Category'
})

df['PM_Ratio'] = df['PM2.5'] / (df['PM10'] + 0.1)


df.to_csv('cleaned_dataset.csv', index=False)

print("\nData cleaning complete. 'cleaned_dataset.csv' has been created.")
```
